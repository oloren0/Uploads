```python
#imports
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Open file, transpose for legibility, then select only the years for which I have other data, 1991-2023, 
#or 2020 as listed here.

healthcare = pd.read_csv('/Users/ofrioren/GIT/Capstone/data/healthcare per state.csv') 
healthcare_df = healthcare.transpose()
healthcare_Cali = healthcare_df[[0, 56]]
healthcare_cali=healthcare_Cali.reset_index()
healthcare_cali
```

```python
#Reformat the file to exclude text explanations.
healthcare_cali=healthcare_cali.drop(healthcare_cali.index[0], axis=0)
healthcare_cali=healthcare_cali.drop(healthcare_cali.index[30], axis=0)
healthcare_cali
```

```python
#Dropping rows again to accomodate only the years for which I have homelessness data. I also changed the coulumn names. 
healthcare_cali=healthcare_cali.drop(healthcare_cali.index[0:16], axis=0)
healthcare_cali=healthcare_cali.rename(columns = {0:'Year', 56:'California'})
healthcare_cali=healthcare_cali.drop(healthcare_cali.columns[0], axis=1)
healthcare_cali
```

```python
#I imported a file with homelessness data, unemployment rates, CPI(consumer price index), housing prices for LA, and the 
#overall homeless count for Los Angeles. I dropped empty columns and 'na' values.
df_los_angeles = pd.read_csv('/Users/ofrioren/GIT/Capstone/data/2007-2023-LA homeless.csv') 
df_los_angeles = df_los_angeles.drop(df_los_angeles.columns[7:], axis=1)
df_los_angeles = df_los_angeles.drop(df_los_angeles.columns[1], axis=1)
df_los_angeles = df_los_angeles.dropna()
df_los_angeles=pd.DataFrame(df_los_angeles)
df_los_angeles
```

```python
#I converted the year into integers instead of floats.
df_los_angeles['Year'] = df_los_angeles['Year'].astype('int')
df_los_angeles
```

```python
#Checked to verify that it converted.
healthcare_cali['Year'] = healthcare_cali['Year'].astype('int')
print(healthcare_cali['Year'].dtype)
```

```python
# I merged the two DataFrames on 'Year' and did some column renaming. 
df_data = pd.merge_ordered(df_los_angeles, healthcare_cali, on='Year')
df_data = df_data.rename(columns = {'California': 'Healthcare Cost, California'})
df_data 
```

```python
#Observable issues: there is also no healthcare cost data for California in years more recent than 2020 through FRED or alternate sources.
```

```python
#I converted data-types in order to delete commas in strings.
df_data['Overall Homeless'] = df_data['Overall Homeless'].astype('str')
df_data['Overall Homeless'] = df_data['Overall Homeless'].str.replace(',', '')
df_data['Overall Homeless'] = df_data['Overall Homeless'].astype('int') 
print(df_data['Overall Homeless'].dtype)
```

```python
#I converted additional data-types in order to delete commas in strings.
df_data['Healthcare Cost, California'] = df_data['Healthcare Cost, California'].fillna(0)
df_data['Healthcare Cost, California'] = df_data['Healthcare Cost, California'].astype('str')
df_data['Healthcare Cost, California'] = df_data['Healthcare Cost, California'].str.replace(',', '')
df_data['Healthcare Cost, California'] = df_data['Healthcare Cost, California'].astype('int') 
print(df_data['Healthcare Cost, California'].dtype)
```

```python
# I created a function, Healthcare_Cost_Incr_Rate, to calculate the percentage increase in healthcare from one year to the next.
def Healthcare_Cost_Incr_Rate(value1, value2):
    rate = ((value2-value1)/(value1)*100)
    return rate
    
HCIR = []
column=df_data['Healthcare Cost, California']
for i in range(0,12):
    rate = Healthcare_Cost_Incr_Rate(column[i], column[i + 1])
    HCIR.append(rate)
    
print(HCIR)
```

```python
#The data above change in value somewhat drastically, so it does not seem appropriate to just use the value from the last year 
#to calculate the increase for the next year. Instead I will use the mean of the values.
HCIR=pd.DataFrame(HCIR)
HCIR.mean()
```

```python
#I used the average health care cost increase to approximate the healthcare cost for the 'na' values from 2020-2023.
base_value = 405451
increment_rate = 1.05094

# Loop through the DataFrame and replace zero values
for i in range(1, len(df_data)):
    if df_data.at[i, 'Healthcare Cost, California'] == 0:
        previous_value = df_data.at[i-1, 'Healthcare Cost, California']
        if previous_value == 0:
            previous_value = base_value
        new_value = previous_value * increment_rate
        df_data.at[i, 'Healthcare Cost, California'] = new_value

df_data
```
