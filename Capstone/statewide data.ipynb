```python
#imports
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#Let's open the statewide data and see what is happening.
state_data = pd.read_excel('/Users/ofrioren/GIT/data/2007-2023-PIT-Counts-by-State.xlsb')
state_data.head()
```

```python
cali_df = state_data[state_data['State'] == 'CA']
cali_df = cali_df.transpose()
cali_df
```

```python
cali_df = cali_df.rename(columns= {5: '% Change in total homelessness statewide'})
cali_df = cali_df.reset_index()
cali_df = cali_df.drop(cali_df.index[0], axis=0)
cali_df
```

```python
cali_df = cali_df.rename(columns= {'index': 'Years'})
cali_df['Years'] = cali_df['Years'].str.replace('Change in Total Homelessness, ', '')
cali_df
```

```python
cali_df['Years'] = cali_df['Years'].str.replace('-2023', '')
cali_df['Years'] = cali_df['Years'].astype('int')
cali_df['Years'].dtype
```

```python
cali_df = cali_df.sort_values(['Years'])
cali_df
```

```python
cali_df_cleaned = cali_df.rename(columns= {'index': 'Year - 2023'})
cali_df_cleaned
```

```python
cali_df_cleaned.to_csv('% cumulative change in homelessness in ca.csv')
```
