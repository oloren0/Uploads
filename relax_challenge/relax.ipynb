```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
import os
import io 

# Show plots inline
%matplotlib inline
```

```python
users = pd.read_csv('/Users/ofrioren/GIT/relax_challenge/takehome_users.csv', encoding='ISO-8859-1')
users
```

```python
user_eng = pd.read_csv('/Users/ofrioren/GIT/relax_challenge/takehome_user_engagement.csv', encoding='ISO-8859-1')
user_eng
```

```python
#Rename a column so I can join the tables
users = users.rename(columns={'object_id':'user_id'})
users
```

```python
#Merge the dfs
user_data = pd.merge(users, user_eng, on='user_id')
user_data
```

```python
# Count the total visits for each user
user_data['total_visits'] = user_data.groupby('user_id')['visited'].transform('count')

# Create the 'adopted' column based on the condition
user_data['adopted'] = user_data['total_visits'].apply(lambda x: 1 if x > 3 else 0)

user_data
```

```python
#What percent of users adopt this product?
adoption = user_data.adopted.sum()/len(user_data['adopted'])
print(adoption)
```

```python
#I'm going to drop some of the columns that do not have predictive value or are repetitions. 
user_data = user_data.drop(['name', 'email', 'last_session_creation_time', 'time_stamp', 'visited'], axis=1)
user_data.info()
```

```python
#I'm going to modify the remaining data to be usable. 

user_data['creation_time'] = pd.to_datetime(user_data['creation_time']).astype('int64') / 1e9

# Perform one-hot encoding on 'creation_source'
ohe = pd.get_dummies(user_data['creation_source'], drop_first=True)

# Concatenate the one-hot encoded columns with the original DataFrame
user_data = pd.concat([user_data, ohe], axis=1).drop(['creation_source'], axis=1)

user_data
```

```python
# Fill NaN values in 'invited_by_user_id' with 0
user_data['invited_by_user_id'] = user_data['invited_by_user_id'].fillna(0)

# Convert the boolean columns to integers (0s and 1s)
booleans = ['ORG_INVITE', 'PERSONAL_PROJECTS', 'SIGNUP', 'SIGNUP_GOOGLE_AUTH']
user_data[booleans] = user_data[booleans].astype(int)

user_data
```

```python
#I'll run an OLS model to see how the features compare.


#imports
import sklearn.model_selection
from sklearn.preprocessing import OneHotEncoder as OHE
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.decomposition import PCA
```

```python
# In order to see how the data collected might indicate adoption, I'll do a Logistic Regression model.

# Split the data into features (X) and target (y)
y = user_data['adopted']
X = user_data.drop(['adopted'], axis=1)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Instantiate and train the LogisticRegression model
lm = LogisticRegression()
lm.fit(X_train, y_train)

# Make predictions using the trained model
predict = lm.predict(X_test)  

# Output table of coefficients per variable
coefficients = pd.DataFrame(lm.coef_[0], X.columns, columns=['Coefficient'])  # lm.coef_ is a 2D array
print(coefficients)
```

```python
#It looks like the most important features are 'user_id', 'creation_time', 'org_id', 'invited-by', and 'total_visits'.
#I'll check another approach just to confirm.

from sklearn.feature_selection import SelectKBest, f_regression

# Assuming X is your feature matrix and y is your target variable
selector = SelectKBest(f_regression, k=5) # Select top 5 features
X_new = selector.fit_transform(X, y)

# Get the selected feature names
selected_features = X.columns[selector.get_support()]
print(selected_features)
```

It appears that the most important factors for user adoption are when they signed up, their organization, the frequency of their visits, which organization invited them, and whether they signed up through the website.
